```python
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nrclex import NRCLex

# Download NLTK data (only once)
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('punkt_tab') # Download the missing corpus

# Sample Data
data = {"Text": [
    "I love this product!",
    "Worst experience ever.",
    "Okay, not bad.",
    "Fantastic!",
    "I hate it."
]}
df = pd.DataFrame(data)

# Clean Text
def clean_text(text):
    text = re.sub(r"http\S+|www\S+", '', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = text.lower()
    stop_words = set(stopwords.words('english'))
    return ' '.join([word for word in text.split() if word not in stop_words])

df['Cleaned'] = df['Text'].apply(clean_text)

# TextBlob Sentiment
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    return 'Positive' if polarity > 0 else 'Negative' if polarity < 0 else 'Neutral'

df['TextBlob_Sentiment'] = df['Cleaned'].apply(get_sentiment)

# VADER Sentiment
sia = SentimentIntensityAnalyzer()
df['VADER_Sentiment'] = df['Text'].apply(
    lambda x: 'Positive' if sia.polarity_scores(x)['compound'] >= 0.05 else
              'Negative' if sia.polarity_scores(x)['compound'] <= -0.05 else
              'Neutral')

# Emotion Analysis (using NRCLex)
def get_emotion(text):
    emotion = NRCLex(text)
    return emotion.top_emotions[0][0] if emotion.top_emotions else 'Unknown'

df['Emotion'] = df['Cleaned'].apply(get_emotion)

# Plot
sns.countplot(data=df, x='VADER_Sentiment')
plt.title("VADER Sentiment Distribution")
plt.show()

# Plot Emotion Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Emotion')
plt.title("Emotion Distribution")
plt.show()
```
